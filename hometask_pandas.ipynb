# **Answer 1**

```python
def answer_one():
    import pandas as pd
    import numpy as np
    # Load Energy Data
    ex = pd.read_excel('Energy Indicators.xls', skiprows=17, index_col=None, skipfooter=38)
    energy = pd.DataFrame(ex)
    energy = energy.drop(['Unnamed: 0', 'Unnamed: 1'], axis=1)      #видаляю перші дві колонки
    energy.rename(columns = {'Unnamed: 2':'Country', 'Petajoules': 'Energy Supply', 'Gigajoules' : 'Energy Supply per Capita', '%': '% Renewable'}, inplace = True)
    energy = energy.replace('...', np.NaN)      #Заміюю пусті значення на Nan'и
    energy['Energy Supply'] = energy['Energy Supply'] * 1000000       #Переводжу з петаджоулі в гігаджулі(multiply=1*10^6)
    energy['Country'].replace('[0-9]+', '', regex=True, inplace=True)     #Щоб забрати верхні індекси біля назв країн
    energy['Country'].replace({'Republic of Korea' : 'South Korea',
                          'United States of America': 'United States',
                          'United Kingdom of Great Britain and Northern Ireland' : 'United Kingdom',
                          'China, Hong Kong Special Administrative Region' : 'Hong Kong'}, inplace=True)
    energy.replace('\([^)]*\)', '', regex=True, inplace=True)
    energy['Country'] = energy['Country'].str.strip()       #забрати відступи
    
    # Load GDP Data
    wb = pd.read_csv('world_bank.csv', skiprows=3, on_bad_lines='skip')
    gdp = pd.DataFrame(wb)
    gdp['Country Name'].replace({'Korea, Rep.' : 'South Korea',
                             'Iran, Islamic Rep.' : 'Iran',
                             'Hong Kong SAR, China' : 'Hong Kong'}, inplace=True)
    gdp = gdp.rename(columns={'Country Name' : 'Country'})
    cols = [ 'Country','2006', '2007', '2008', '2009', '2010', '2011',
        '2012', '2013', '2014', '2015']
    gdp_merge = gdp[cols]
    
    # Load Country Rank data
    file_name = 'scimagojr.xlsx'
    sc = pd.read_excel(file_name)
    ScimEn = pd.DataFrame(sc)
    ScimEn = ScimEn[ScimEn['Rank'] <= 15]

    
    df_merge = pd.merge(pd.merge(energy, ScimEn, how='inner', on='Country'), gdp_merge, how='inner', on='Country')   
    df_merge.set_index('Country', inplace=True)   #виставляю поле Country як індексне поле
    return df_merge
    
    
answer_one()
```

# **Answer 2**

```python
def answer_two():
    Top15 = answer_one()
    Top15['avgGDP'] = answer_one()[cols[1:]].mean(axis=1)
    return Top15.sort_values('avgGDP',ascending=False)['avgGDP']
    

answer_two()

```

# **Answer 3**

```python
def answer_three():
    Top15 = answer_one()
    Top15['avgGDP'] = answer_two()
    Top15.sort_values('avgGDP', ascending=False, inplace=True)
    start = Top15.iloc[5]['2006']
    stop = Top15.iloc[5]['2015']
    return abs(stop - start)
    
answer_three()
```

```python
def answer_five():
    Top15 = answer_one()
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    srt = Top15.sort_values('Population',ascending = False)
    #return srt['Population'].max()
    return srt.iloc[2].name

answer_five()
```

# **Answer 6**

```python
def answer_six():
    Top15 = answer_one()
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable documents per person'] = Top15['Citable documents'] / Top15['Population']
    return Top15['Citable documents per person'].corr(Top15['Energy Supply per Capita'])

answer_six()
```

# **Answer 7**


```python
def answer_seven():
    Top15 = answer_one()
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    df_groups = pd.DataFrame(columns=['size', 'sum', 'mean', 'std'])
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    dictionary = Top15.groupby(ContinentDict)
    for group, frame in dictionary:
        df_groups.loc[group] = [len(frame), frame['Population'].sum(),
                                            frame['Population'].mean(), 
                                            frame['Population'].std()]
    return df_groups

answer_seven()
```
